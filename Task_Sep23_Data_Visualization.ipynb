```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

Task 1: Basic Analysis

```python
#Total Revenue
total_revenue = df.groupby("Product")["Revenue"].sum()
print("Total Revenue per Product:\n ", total_revenue)
```

```python
# average units sold per product
avg_units_sold = df.groupby("Product")["Units_Sold"].mean()
print("Average Units sold per Product:\n ", avg_units_sold)
```

Task 2: Visualization - Line chart of Monthly Revenue Trend

```python
import matplotlib.pyplot as plt
# plt.plot(df[df["Product"] == "Mobile"]["Month"],df[df["Product"] == "Mobile"]["Revenue"],marker='o',label="Mobile")
# plt.plot(df[df["Product"] == "Laptop"]["Month"],df[df["Product"] == "Laptop"]["Revenue"],marker='o',label="Laptop")
# plt.plot(df[df["Product"] == "TV"]["Month"],df[df["Product"] == "TV"]["Revenue"],marker='o',label="TV")

plt.figure(figsize = (8,6))
for product in df["Product"].unique():
    month_data = df[df["Product"] == product]
    plt.plot(month_data["Month"], month_data["Revenue"],marker = "o",label = product)
plt.title("Monthly Revenue Trend per Product")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

Task 3: Visualization - Bar Chart

```python
plt.figure(figsize = (8,6))
total_revenue.plot(kind='bar',color = ['red','blue','green'])
plt.title("Total Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

Task 4: Visualization - Pie Chart

```python
plt.figure(figsize = (8,6))
total_revenue.plot(kind='pie',color = ['red','blue','green'])
# total_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Revenue contribution per Product")
plt.ylabel("")
plt.show()
```
